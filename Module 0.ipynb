#Feature Engineering

fitness feature endurance

Combine sit-and-reach (left + right)
Combine grip strength (dominant + non-dominant)
Convert endurance time → total seconds


```python
# ===============================
# CELL 11 — Fitness Feature Engineering
# ===============================

# --- Sit and Reach (average of left + right)
df["FGC_SR_avg"] = df[["FGC-FGC_SRL", "FGC-FGC_SRR"]].mean(axis=1)

# --- Grip Strength (average of dominant + non-dominant)
df["FGC_GS_avg"] = df[["FGC-FGC_GSD", "FGC-FGC_GSND"]].mean(axis=1)

# --- Endurance total time in seconds
df["Endurance_Time_Total"] = (
    df["Fitness_Endurance-Time_Mins"] * 60 +
    df["Fitness_Endurance-Time_Sec"]
)

print("Feature engineering for fitness completed.")
```

```python
# ===============================
# CELL 12 — Validate Engineered Features
# ===============================

df[[
    "FGC_SR_avg",
    "FGC_GS_avg",
    "Endurance_Time_Total"
]].head()
```

```python
# ===============================
# CELL 13 — PAQ Feature Engineering
# ===============================

# Combine PAQ_A and PAQ_C into one unified PAQ score
df["PAQ_Total"] = df["PAQ_A-PAQ_A_Total"].combine_first(df["PAQ_C-PAQ_C_Total"])

# Quick check
df[["PAQ_A-PAQ_A_Total", "PAQ_C-PAQ_C_Total", "PAQ_Total"]].head()
```

```python
# ===============================
# CELL 15 — Block-Level Indicators
# ===============================

import numpy as np

# SDS
df["SDS_measured"] = np.where(
    df[["SDS-SDS_Total_Raw", "SDS-SDS_Total_T"]].notna().any(axis=1), 1, 0
)

# PAQ
df["PAQ_measured"] = np.where(
    df[["PAQ_Total"]].notna().any(axis=1), 1, 0
)

# Fitness
df["Fitness_measured"] = np.where(
    df[["Fitness_Endurance-Max_Stage", "Endurance_Time_Total"]].notna().any(axis=1), 1, 0
)

# BIA block
bia_cols = [
    "BIA-BIA_Activity_Level_num", "BIA-BIA_BMC", "BIA-BIA_BMR", "BIA-BIA_DEE",
    "BIA-BIA_ECW", "BIA-BIA_FFM", "BIA-BIA_FFMI", "BIA-BIA_FMI",
    "BIA-BIA_Fat", "BIA-BIA_Frame_num", "BIA-BIA_ICW", "BIA-BIA_LDM",
    "BIA-BIA_LST", "BIA-BIA_SMM", "BIA-BIA_TBW"
]

df["BIA_measured"] = np.where(df[bia_cols].notna().any(axis=1), 1, 0)

# FGC block
fgc_cols = [
    "FGC-FGC_CU", "FGC-FGC_CU_Zone", "FGC-FGC_PU", "FGC-FGC_PU_Zone",
    "FGC-FGC_TL", "FGC-FGC_TL_Zone",
    "FGC-FGC_GSD_Zone", "FGC-FGC_GSND_Zone",
    "FGC-FGC_SRL_Zone", "FGC-FGC_SRR_Zone",
    "FGC_SR_avg", "FGC_GS_avg"
]

df["FGC_measured"] = np.where(df[fgc_cols].notna().any(axis=1), 1, 0)

print("Block-level indicators created.")

df[[
    "SDS_measured",
    "PAQ_measured",
    "Fitness_measured",
    "BIA_measured",
    "FGC_measured"
]].head()
```

##Block wise KNN imputation

```python
# ===============================
# CELL 21 — Clean Physical Values
# ===============================

# Height
df.loc[(df["Physical-Height"] < 50) | (df["Physical-Height"] > 250), "Physical-Height"] = np.nan

# Weight
df.loc[(df["Physical-Weight"] < 10) | (df["Physical-Weight"] > 300), "Physical-Weight"] = np.nan

# BMI
df.loc[(df["Physical-BMI"] < 10) | (df["Physical-BMI"] > 60), "Physical-BMI"] = np.nan

# Blood Pressure
df.loc[(df["Physical-Systolic_BP"] < 70) | (df["Physical-Systolic_BP"] > 250), "Physical-Systolic_BP"] = np.nan
df.loc[(df["Physical-Diastolic_BP"] < 40) | (df["Physical-Diastolic_BP"] > 150), "Physical-Diastolic_BP"] = np.nan

# Heart Rate
df.loc[(df["Physical-HeartRate"] < 30) | (df["Physical-HeartRate"] > 220), "Physical-HeartRate"] = np.nan

print("Physical values cleaned.")
```

```python
# ===============================
# CELL 22 — Physical KNN Imputation
# ===============================

from sklearn.impute import KNNImputer

physical_cols = [
    "Physical-Height",
    "Physical-Weight",
    "Physical-BMI",
    "Physical-Systolic_BP",
    "Physical-Diastolic_BP",
    "Physical-HeartRate"
]

imputer = KNNImputer(n_neighbors=5)

df[physical_cols] = imputer.fit_transform(df[physical_cols])

print("Physical KNN imputation done.")
df[physical_cols].isnull().sum()
```

BIA KNN imputation

```python
# ===== FINAL BIA COLUMNS =====
bia_selected_knn = [
    "BIA-BIA_Fat",
    "BIA-BIA_FMI",
    "BIA-BIA_FFMI",
    "BIA-BIA_SMM"
]

bia_extra_model = [
    "BIA-BIA_FFM"
]

bia_other_keep = [
    "BIA-BIA_Activity_Level_num",
    "BIA-BIA_Frame_num"
]

bia_all_final = bia_selected_knn + bia_extra_model + bia_other_keep

print("BIA columns to impute:")
print(bia_all_final)
```

```python
# ===== MISSING BEFORE IMPUTATION =====
bia_missing_before = df[bia_all_final].isnull().sum().sort_values(ascending=False)
print(bia_missing_before)
```

```python
# ===== CLEAN INVALID VALUES BEFORE KNN =====

# fat %
df.loc[(df["BIA-BIA_Fat"] < 0) | (df["BIA-BIA_Fat"] > 60), "BIA-BIA_Fat"] = np.nan

# fat mass index
df.loc[(df["BIA-BIA_FMI"] < 0) | (df["BIA-BIA_FMI"] > 30), "BIA-BIA_FMI"] = np.nan

# fat-free mass index
df.loc[(df["BIA-BIA_FFMI"] < 5) | (df["BIA-BIA_FFMI"] > 30), "BIA-BIA_FFMI"] = np.nan

# skeletal muscle mass
df.loc[(df["BIA-BIA_SMM"] < 5) | (df["BIA-BIA_SMM"] > 100), "BIA-BIA_SMM"] = np.nan

# fat-free mass
df.loc[(df["BIA-BIA_FFM"] < 10) | (df["BIA-BIA_FFM"] > 150), "BIA-BIA_FFM"] = np.nan

# activity level category
df.loc[(df["BIA-BIA_Activity_Level_num"] < 1) | (df["BIA-BIA_Activity_Level_num"] > 5), "BIA-BIA_Activity_Level_num"] = np.nan

# frame category
df.loc[(df["BIA-BIA_Frame_num"] < 1) | (df["BIA-BIA_Frame_num"] > 3), "BIA-BIA_Frame_num"] = np.nan
```

```python
# ===== CLIP EXTREME VALUES =====
for col in bia_selected_knn + bia_extra_model:
    lower = df[col].quantile(0.01)
    upper = df[col].quantile(0.99)
    df[col] = df[col].clip(lower, upper)
```

```python
# ╔════════════════════════════════════════════════════════════════════╗
# ║         FEATURE ENGINEERING — BEHAVIORAL / METABOLIC              ║
# ╚════════════════════════════════════════════════════════════════════╝

import numpy as np

# -------------------------------------------------------------------
# 1. Social_Digital_Displacement
# -------------------------------------------------------------------
# Idea:
# More screen time + lower CGAS = more digital/social displacement.
# +1 in denominator avoids division by zero.
df['Social_Digital_Displacement'] = (
    df['PreInt_EduHx-computerinternet_hoursday'] /
    (df['CGAS-CGAS_Score'] / 10 + 1)
)

# -------------------------------------------------------------------
# 2. Lifestyle_Risk_Score
# -------------------------------------------------------------------
# Combines:
# - higher SDS = more distress
# - lower PAQ = less activity
# - higher screen time = more sedentary behavior

df['Lifestyle_Risk_Score'] = (
    df['SDS-SDS_Total_T'] / 100 +
    1 / (df['PAQ_Total'] + 1) +
    df['PreInt_EduHx-computerinternet_hoursday'] / 10
)

# -------------------------------------------------------------------
# 3. Metabolic_Alarms_Count
# -------------------------------------------------------------------
# Counts how many risk alarms are triggered.

bmi_alarm = (df['Physical-BMI'] > 25).astype(int)
bp_alarm = (df['Physical-Systolic_BP'] > 120).astype(int)
sedentary_alarm = (df['PAQ_Total'] < 2.0).astype(int)
fat_alarm = (df['BIA-BIA_Fat'] > 30).astype(int)
fitness_alarm = (df['Fitness_Endurance-Max_Stage'] < 3).astype(int)

df['Metabolic_Alarms_Count'] = (
    bmi_alarm +
    bp_alarm +
    sedentary_alarm +
    fat_alarm +
    fitness_alarm
)

# -------------------------------------------------------------------
# QUICK CHECK
# -------------------------------------------------------------------
new_features = [
    'Social_Digital_Displacement',
    'Lifestyle_Risk_Score',
    'Metabolic_Alarms_Count'
]

print("✅ New features created successfully.\n")

print("Missing values:")
print(df[new_features].isnull().sum())

print("\nDescriptive statistics:")
print(df[new_features].describe())

print("\nMetabolic_Alarms_Count distribution:")
print(df['Metabolic_Alarms_Count'].value_counts().sort_index())
# ╔════════════════════════════════════════════════════════════════════╗
# ║        ADDITIONAL SCREEN TIME FEATURES                            ║
# ╚════════════════════════════════════════════════════════════════════╝

# -------------------------------------------------------------------
# 4. ScreenTime_Age_Ratio
# -------------------------------------------------------------------
# Captures how much screen time relative to age (younger + high screen = worse)

df['ScreenTime_Age_Ratio'] = (
    df['PreInt_EduHx-computerinternet_hoursday'] /
    (df['Basic_Demos-Age'] + 1)   # +1 to avoid division by zero
)

# -------------------------------------------------------------------
# 5. ScreenTime_Category (ordinal feature)
# -------------------------------------------------------------------

def screen_cat(x):
    if x < 1:
        return 0
    elif x < 2.5:
        return 1
    elif x < 4:
        return 2
    else:
        return 3

df['ScreenTime_Category'] = df['PreInt_EduHx-computerinternet_hoursday'].apply(screen_cat)

# -------------------------------------------------------------------
# QUICK CHECK
# -------------------------------------------------------------------

new_cols = ['ScreenTime_Age_Ratio', 'ScreenTime_Category']

print("✅ Screen time features added.\n")

print("Missing values:")
print(df[new_cols].isnull().sum())

print("\nDistribution:")
for col in new_cols:
    print(f"\n{col}")
    print(df[col].value_counts().sort_index() if col.endswith('Category') else df[col].describe())
```
